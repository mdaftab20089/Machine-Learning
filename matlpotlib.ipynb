## Matplotlib Basics

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x=np.arange(0,10)
y=np.arange(23,33)
a=np.arange(40,50)
b=np.arange(60,70)

plt.scatter(x,y,c='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D - Graph')
plt.plot(x,y,'r*-')
```

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2D graph')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,1.414,1.732,2,2.23606797749979]
plt.plot(x,y,color='red',linestyle='--',marker='*',linewidth=3,markersize=9)
plt.grid(True) ## its not neccessary
plt.show()
```

```python
## subplots of a plot
x1=[1,2,3,4,5]
y1=[1,1.4142135623730951,1.7320508075688772,2,2.23606797749979]
x2=[1,2,3,4,5]
y2=[1,8,27,64,125]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x1,y1,color='red')
plt.title('PLOT-1')

plt.subplot(2,2,2)
plt.plot(x1,y2,color='green')
plt.title('PLOT-2')

plt.subplot(2,2,3)
plt.plot(x2,y1,color='black')
plt.title('PLOT-3')

plt.subplot(2,2,4)
plt.plot(x2,y2,color='pink')
plt.title('PLOT-4')
```

## Bar Plot

```python
categories=['a','b','c','d','e']
values=[1,2,3,4,5]

## create a bar plot
plt.bar(categories,values,color='purple')
```

## Histogram- it is used to represent the distribution of data they divide the data into the beans and count the no of points in each beans.

```python
data =[1,2,3,4,2,3,4,1,1,6,3,4,5,6,3]

## creata histogram
plt.hist(data,bins=6)
```

```python
plt.hist(data,bins=6,color='orange',edgecolor='pink')
```

```python
## scatter plot
x2=[1,2,3,4,5]
y2=[1,8,27,64,125]
plt.scatter(x2,y2,color='blue',marker='x')
```

```python
## Bar plot
labels=['a','b','c','d']
sizes=[2,3,4,1]
#colors=['gold','yellowgreen','lightcoral','lightskyblue']
#explode=(0.1,0,0,0) ##move out the 1st slice
plt.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True)
```
