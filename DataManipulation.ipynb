## Merging And Joining Data Frames

```python
df1=pd.DataFrame({'key':['a','b','c'] , 'value':[1,2,3]})
df2=pd.DataFrame({'key':['a','b','d'] , 'value':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# merge data frame on the behalf of column 1
pd.merge(df1,df2,on="key",how="inner") ## consider only the commonn data presesnt in the data frames
```

```python
pd.merge(df1,df2,on="key",how="outer") ## consider all data present in the data frames
```

```python
pd.merge(df1,df2,on="key",how="left") ## giving more priority for the left data
```

```python
pd.merge(df1,df2,on="key",how="right") ## giving more priority for the right data
```
