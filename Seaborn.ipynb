## Seaborn in Python for data visualisation

```python
import seaborn as sns
```

```python
## basic plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot of total bill vs tips')
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total bill vs tips')
plt.show()
```

```python
## categorical plot in seaborn
## bar plot
sns.barplot(x='total_bill',y='tip',data=tips)
```

```python
## violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## histogram under seaborn
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## kde plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## pair plot
sns.pairplot(tips)
```

```python
## heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
