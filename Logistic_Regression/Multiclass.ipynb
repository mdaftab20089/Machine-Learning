```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
## building the model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, GridSearchCV

# Define the correct hyperparameters for Logistic Regression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'solver': ['saga', 'liblinear'],
    'max_iter': [100, 200, 300]
}

# Initialize Logistic Regression model
model = LogisticRegression(multi_class='ovr')

# Stratified K-Fold Cross Validation
cv = StratifiedKFold(n_splits=5)

# Grid Search for Hyperparameter Tuning
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=cv)
grid.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Hyperparameters:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```
